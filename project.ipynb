```python
try:
    connection = mysql.connector.connect(
        host="localhost",
        user="student",
        password='0TOGXxAgHY1HX',
        database="student"# Replace with the actual password
    )

    if connection.is_connected():
        print("Connected to MySQL")

        # Create a cursor
        cursor = connection.cursor()

        # Execute a query to create stock_merged table
        query = '''create table stock_merged(
        SELECT 
            s1.Time,
            s1.Close AS Close_CNQ,
            s2.Close AS Close_SU,
            s3.Close AS Close_CVE,
            s4.Close AS Close_IMO
        FROM stock_CNQ s1
        LEFT JOIN stock_SU s2 ON s2.Time = s1.Time
        LEFT JOIN stock_CVE s3 ON s3.Time = s1.Time
        LEFT JOIN stock_IMO s4 ON s4.Time = s1.Time
        ORDER BY s1.Time);'''
        cursor.execute(query)
except Error as e:
    print("Error while connecting to MySQL:", e)

finally:
    # Ensure the cursor and connection are closed properly
    if connection and connection.is_connected():
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
```

```python
import pandas as pd

# Load the data
df_prices = pd.read_csv("Oil Prices.csv")

# Display initial data & check for missing values
print(df_prices.head())
print(df_prices.isnull().sum())

# Ensure 'Date' column is string
df_prices['Date'] = df_prices['Date'].astype(str)

# Extract last two digits (YY) and convert to integers
df_prices['Year'] = df_prices['Date'].str[-2:].astype(int)

# Convert two-digit year to four-digit format
df_prices['Year'] = df_prices['Year'].apply(lambda x: 1900 + x if x >= 86 else 2000 + x)

# Replace old year in 'Date' with the new fixed year
df_prices['Date'] = df_prices['Date'].str[:-2] + df_prices['Year'].astype(str)

# Convert to datetime
df_prices['Date'] = pd.to_datetime(df_prices['Date'])

# Rename columns
df_prices.rename(columns={'Date': 'Time', 'Value': 'price', 'Series': 'series', 'labels': 'Record_time'}, inplace=True)

# Drop the unnecessary column
df_prices.drop(columns=['Record_time', 'Year'], inplace=True)
df_prices=df_prices[(df_prices['Time'] >= "2000-01-01") & (df_prices['Time'] <= "2024-12-01")]
# Display final dataset
print(df_prices.head())
print(len(df_prices))
```

```python
trade_rank_query='''WITH Combined_Trades AS (
    SELECT  "CNQ" AS company ,volume, Time FROM stock_CNQ
    UNION ALL
    SELECT  "CVE" AS company, volume, Time FROM stock_CVE
    UNION ALL
    SELECT  "SUTO" AS company,volume, Time FROM stock_SU
    UNION ALL
    SELECT "IMO" AS company ,volume, Time FROM stock_IMO
),
Total_Trade_Volume AS (
    SELECT company,
        SUM(volume) AS total_volume
    FROM Combined_Trades
    GROUP BY company
)
SELECT 
    company, 
    total_volume,
    RANK() OVER (ORDER BY total_volume DESC) AS trade_rank
FROM Total_Trade_Volume;

'''
df= pd.read_sql_query(trade_rank_query, myconnection)
df.head()
```

```python
volatility_query='''WITH Combined_Trades AS (
    SELECT  "CNQ" AS company ,((close - open) / open) * 100 AS pct_change, Time FROM stock_CNQ
    UNION ALL
    SELECT  "CVE" AS company, ((close - open) / open) * 100 AS pct_change, Time FROM stock_CVE
    UNION ALL
    SELECT  "SUTO" AS company,((close - open) / open) * 100 AS pct_change, Time FROM stock_SU
    UNION ALL
    SELECT "IMO" AS company ,((close - open) / open) * 100 AS pct_change, Time FROM stock_IMO
),
volatility AS (
    SELECT company,
        stddev(pct_change) AS volatility
    FROM Combined_Trades
    GROUP BY company
)
SELECT 
    company, 
    volatility,
    RANK() OVER (ORDER BY volatility DESC) AS Volatility_rank
FROM volatility;

'''
df= pd.read_sql_query(volatility_query, myconnection)
df.head()
```

```python
close_rank_query='''WITH Combined_Trades AS (
     SELECT  "CNQ" AS company ,Close, Time FROM stock_CNQ
    UNION ALL
    SELECT  "CVE" AS company, Close, Time FROM stock_CVE
    UNION ALL
    SELECT  "SUTO" AS company,Close, Time FROM stock_SU
    UNION ALL
    SELECT "IMO" AS company ,Close, Time FROM stock_IMO
),
Total_Trade_Volume AS (
    SELECT company,
        avg(close) AS Avg_close
    FROM Combined_Trades
    GROUP BY company
)
SELECT 
    company, 
    Avg_close,
    RANK() OVER (ORDER BY Avg_close DESC) AS Close_rank
FROM Total_Trade_Volume;
'''
df= pd.read_sql_query(close_rank_query, myconnection)
df.head()
```

```python
yoy_growth_query = """
WITH YearlyStock AS (
    SELECT 
        YEAR(Time) AS year,
        'CNQ' AS stock_name, 
        AVG(Close_CNQ) AS avg_close
    FROM stock_merged
    GROUP BY year
    UNION ALL
    SELECT 
        YEAR(Time) AS year,
        'CVE' AS stock_name, 
        AVG(Close_CVE) AS avg_close
    FROM stock_merged
    GROUP BY year
    UNION ALL
    SELECT 
        YEAR(Time) AS year,
        'SU' AS stock_name, 
        AVG(Close_SU) AS avg_close
    FROM stock_merged
    GROUP BY year
    UNION ALL
    SELECT 
        YEAR(Time) AS year,
        'IMO' AS stock_name, 
        AVG(Close_IMO) AS avg_close
    FROM stock_merged
    GROUP BY year
),
StockGrowth AS (
    SELECT 
        stock_name,
        (MAX(avg_close) - MIN(avg_close)) / MIN(avg_close) * 100 AS total_growth
    FROM YearlyStock
    GROUP BY stock_name
),
RankedStocks AS (
    SELECT 
        stock_name,
        total_growth,
        RANK() OVER (ORDER BY total_growth DESC) AS overall_rank
    FROM StockGrowth
)
SELECT * FROM RankedStocks
ORDER BY overall_rank;
"""
df_yoy_growth = pd.read_sql(yoy_growth_query, myconnection)
df_yoy_growth.head()
```

```python
# Pie chart plot for oil type crude oil exports
oil_type_query='''WITH OilTotals AS (
    SELECT 
        Oil_Type, 
        SUM(`Volume_(m3/d)`) AS total_volume
    FROM exports
    WHERE Oil_Type != 'Total'  -- Exclude 'Total' row if it exists
    GROUP BY Oil_Type
),
GrandTotal AS (
    SELECT SUM(total_volume) AS grand_total FROM OilTotals
)
SELECT 
    O.Oil_Type, 
    O.total_volume, 
    (O.total_volume / G.grand_total) * 100 AS proportion_percentage
FROM OilTotals O
CROSS JOIN GrandTotal G
ORDER BY proportion_percentage DESC;
'''

df_export_type= pd.read_sql_query(oil_type_query, myconnection)
plt.figure(figsize=(8, 8))
plt.pie(df_export_type["proportion_percentage"], labels=df_export_type["Oil_Type"], autopct='%1.1f%%', colors=["gold", "blue", "red", "green"])
plt.title("Proportion of Oil Types in Total Production")
plt.show()
```
